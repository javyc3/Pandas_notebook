  ## Create DataFrame For Gender

```python
# count the unique values of the gender column 
gender_count = df["Gender"].value_counts() 

# create dataframe from unique value count of the gender column
gender_demograph = pd.DataFrame(gender_count) 

# Rename Gender column 
gender_demograph = gender_demograph.rename(columns = {'Gender':'Total Count' }) 

# change the 'total count' column to numeric value
gender_demograph['Total Count'] = pd.to_numeric(gender_demograph['Total Count']) 

# find the sum of 'total count' column
total_gd = gender_demograph['Total Count'].sum() 

# percentage of each gender in the column
percentage_of_players = gender_demograph['Total Count'] / total_gd 

# add the percentage of players to the dataframe
gender_demograph['Percentage of Players'] = percentage_of_players 

gender_demograph  
```

```python
# group data by gender 
gender_analysis = df.groupby(['Gender']) 

# count the number of purchases by gender
pur_count = gender_analysis['Price'].count()

# find the average purchase amount by gender
purchase_avg = gender_analysis['Price'].mean() 

# find the total purchase value of each gender
pur_val_sum = gender_analysis['Price'].sum()  

# create a sumary table for gender data
gender_summary = pd.DataFrame({'Purchase Count': pur_count, 
                  'Average Purchase Price': purchase_avg, 
                  'Total Purchase Value': pur_val_sum}) 

# add 'percentage of players' column to the dataframe
gender_summary['Percentage of Players'] = percentage_of_players 

gender_summary
```

## Create Dataframe for the top Purchasers

```python
# group the data by players name 
top_spender = df.groupby('SN') 

# count the number of purchases of each player
purchase_count = top_spender['Item ID'].count() 

# get the sum each players purchases
total_purchase = top_spender['Price'].sum() 

# Average price of each purchase by each player
average_purchase_price = top_spender['Price'].sum()/purchase_count 

# Create dataframe for top Purchasers 
top_spender_summary = pd.DataFrame({'Purchase Count': purchase_count, 'Average Purchase Price': average_purchase_price, 
                       'Total Purchase Value': total_purchase}) 

# sort values in descending order
top_spender_summary.sort_values('Total Purchase Value', ascending = False).head(5)
```

## Create Dataframe for  top 5 Popular items

```python
# Group dataFrame by 'item name' and 'item ID'
pop_items = df.groupby(['Item Name', "Item ID"]) #183 rows

# Count the number of items in the Dataframe
pop_items.count() 

# Get a price list of the items 
item_price = pop_items['Price'] 

# The total price of all items in the list
total_purchase_value = pop_items['Price'].sum() 

# Count the purchases made for each item 
purchase_count = pop_items['Item ID'].count()

# Create dataframe for items
most_pop_summary = pd.DataFrame({'Purchase Count': purchase_count, 'Item Price':item_price, 
                                 'Total Purchase Value': total_purchase_value }) 

# sort data by in descending order and print top 5 items 
most_pop_summary = most_pop_summary.sort_values('Purchase Count', ascending = False).head(5) 
most_pop_summary
```
